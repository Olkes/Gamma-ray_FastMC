```python
import sys
sys.path.insert(0, "../python/")

import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
%matplotlib inline
```

Load the .json file with the Monte Carlo result

```python
#df = pd.read_json('../mcdata/testdata.json').dropna()
df= pd.read_csv('../mcdata/testdata6.csv',names=['#','nscatters','w','de','x0','y0','z0','x1','y1','z1','de1','x2','y2','z2','de2','x3','y3','z3','de3','x4','y4','z4','de4'], error_bad_lines=False)
df.shape
df_fmc= pd.read_csv('../mcdata/RMS_test_data2/testdata1000000_1.csv',names=['#','nscatters','w','de','x0','y0','z0','x1','y1','z1','de1','x2','y2','z2','de2','x3','y3','z3','de3','x4','y4','z4','de4'], error_bad_lines=False)
df.shape
```

Plot the hit position

```python
plt.figure(figsize=(11,5))
plt.subplot(1, 2, 1)
h=plt.hist2d(df['x1'],df['y1'],bins=(100,100),weights=df['w'],range=[[-55,55],[-55,55]],norm=LogNorm())
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.subplot(1,2,2)

r2 = df['x1']**2+df['y1']**2
h=plt.hist2d(r2,df['z1'],bins=(101,101),weights=df['w'],range=[[0,2500],[-55,55]],norm=LogNorm())
plt.xlabel('$R^2$ (cm$^2$)')
plt.ylabel('z (cm)');
```

```python
df1 = df[(df['de']<250) & (r2<1600) & (np.abs(df['z1'])<40) & (df['nscatters']==1)]  
```

```python
plt.figure(figsize=(11,5))
plt.subplot(1, 2, 1)
h=plt.hist2d(df1['x1'],df1['y1'],bins=(100,100),weights=df1['w'],range=[[-55,55],[-55,55]],norm=LogNorm())
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.subplot(1,2,2)

r22 = df1['x1']**2+df1['y1']**2
h=plt.hist2d(r22,df1['z1'],bins=(101,101),weights=df1['w'],range=[[0,2500],[-55,55]],norm=LogNorm())
plt.xlabel('$R^2$ (cm$^2$)')
plt.ylabel('z (cm)');
```
